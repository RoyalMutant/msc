```python
import cv2
import numpy as np

# Load a grayscale image
img = cv2.imread("D:\\Code\\cat.jpeg", cv2.IMREAD_GRAYSCALE)

# Ensure the image is 512x512
img = cv2.resize(img, (512, 512))

subsampled_imgs = [img]

# Subsampling
for i in range(4):
    img = img[::2, ::2]
    subsampled_imgs.append(img)

# Resampling
resampled_imgs = []

for i in subsampled_imgs[1:]:
    resampled = i.copy()
    while resampled.shape[0] < 512:
        resampled = np.repeat(resampled, 2, axis=0)
        resampled = np.repeat(resampled, 2, axis=1)
    resampled_imgs.append(resampled)

titles = ["256x256 -> 512x512", "128x128 -> 512x512", "64x64 -> 512x512", "32x32 -> 512x512"]

# Display resampled images
for i in range(4):
    cv2.imshow(titles[i], resampled_imgs[i])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load grayscale image
img = cv2.imread("D:\\Code\\cat.jpeg", cv2.IMREAD_GRAYSCALE)

# Resize to 512x512 using cv2 just once
img = cv2.resize(img, (512, 512))

# Manual subsampling
subsampled_imgs = [img]
for _ in range(4):
    img = img[::2, ::2]
    subsampled_imgs.append(img)

# Manual resampling (nearest-neighbor upsampling without np.repeat)
resampled_imgs = []

for down in subsampled_imgs[1:]:  # Skip the original
    h, w = down.shape
    factor = 512 // h  # how much we need to upscale

    # Create an empty array of shape (512, 512)
    upsampled = np.zeros((512, 512), dtype=np.uint8)

    # Fill in each pixel by repeating the original pixels
    for i in range(h):
        for j in range(w):
            value = down[i, j]
            for x in range(i * factor, (i + 1) * factor):
                for y in range(j * factor, (j + 1) * factor):
                    upsampled[x, y] = value

    resampled_imgs.append(upsampled)

# Show images
titles = ["256x256 -> 512x512", "128x128 -> 512x512", "64x64 -> 512x512", "32x32 -> 512x512"]
for i in range(4):
    cv2.imshow(titles[i], resampled_imgs[i])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
