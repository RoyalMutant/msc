```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_img(img_path):
    """Loads a grayscale image."""
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    return img

def modify_lsb(img):
    modified_img = img - (img % 2)
    return modified_img

def save_img(img, filename):
    cv2.imwrite(filename, img)

def display_imgs(original, modified, difference):
    imgs = [original, modified, difference]
    titles = ['Original image', 'Modified image', 'Difference image']
    
    plt.figure(figsize=(12, 4))
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(imgs[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    
    plt.show()

def compute_difference(original, modified):
    return original - modified

def analyze_difference(difference):
    print("Min Difference:", np.min(difference))
    print("Max Difference:", np.max(difference))
    print("Mean Difference:", np.mean(difference))

original = load_img("D:\\Code\\lenna.jpeg")
modified = modify_lsb(original)
save_img(modified, "modified_image.jpg")
difference = compute_difference(original, modified)
display_imgs(original, modified, difference)
analyze_difference(difference)
```

```python
import cv2
import numpy as np

def load_img_grayscale(img_path):
    return cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

def convert_to_float(img):
    return img.astype(np.float32)

def get_min_max(img):
    min_val = np.min(img)
    max_val = np.max(img)
    return min_val, max_val

def normalize_img(img, R=128):
    min_val, max_val = get_min_max(img)
    if max_val == min_val:
        return np.zeros(img, dtype=np.uint8)
    norm = (img - min_val) / (max_val - min_val) * R
    return norm.astype(np.uint8)

def display_imgs(original, norm):
    cv2.imshow("Original image", original)
    cv2.imshow("Normalized image", norm)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

original_img = load_img_grayscale("D:\\Code\\lenna.jpeg")
img_float = convert_to_float(original_img)
norm_img = normalize_img(img_float, R=128)
display_imgs(original_img, norm_img)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_img(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    return img

def generate_key(img_shape):
    return np.random.randint(0, 256, img_shape)

def encrypt_img(img, key):
    encrypted_img = (img + key) % 256
    return encrypted_img

def decrypt_img(encrypted_img, key):
    decrypted_img = (encrypted_img - key) % 256
    return decrypted_img

def display_imgs(original, encrypted, decrypted):
    imgs = [original, encrypted, decrypted]
    titles = ["Original image", "Encrypted image", "Decrypted image"]
    
    plt.figure(figsize=(12, 4))
    for i in range(3):
        plt.subplot(1, 3, i + 1)
        plt.title(titles[i])
        plt.imshow(imgs[i], cmap='gray')
        plt.axis("off")
    
    plt.show()

original_img = load_img("D:\\Code\\lenna.jpeg")
key_matrix = generate_key(original_img.shape)
encrypted_img = encrypt_img(original_img, key_matrix)
decrypted_img = decrypt_img(encrypted_img, key_matrix)
display_imgs(original_img, encrypted_img, decrypted_img)
```
